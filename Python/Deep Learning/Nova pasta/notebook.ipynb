```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
```

```python
mpl.rcParams['figure.figsize'] = (10, 6)
mpl.rcParams['font.size'] = (12)
```

```python
url = "https://raw.githubusercontent.com/alura-cursos/deeptime/aula1/Passageiros.csv"
df = pd.read_csv(url)
```

```python
df.head()
df.shape
```

```python
sns.lineplot(data = df, x = 'tempo', y = 'passageiros') 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df)
df_escalado = scaler.transform(df)
```

```python
x = df_escalado[:, 0]
y = df_escalado[:, 1]

qtd_treino = int(df_escalado.shape[0] * 0.75)

x_treino = x[0:qtd_treino]
x_teste = x[qtd_treino:]
y_treino =  y[0:qtd_treino]
y_teste = y[qtd_treino:]
```

```python
sns.lineplot(x = x_treino, y = y_treino, label = 'treino')
sns.lineplot(x = x_teste, y = y_teste, label = 'teste')
```
