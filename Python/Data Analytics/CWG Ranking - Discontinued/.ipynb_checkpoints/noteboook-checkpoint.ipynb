## Differences Between good and bad wine

### What makes a wine good?
* Good wine has less volatile acidiy (50% less compared to bad wine)
* It has more citric acid than bad wine (+200% more when compared)
* A LOT less chlorides (50% less compared to bad wine)
* More Sulphats (about 34% more than bad wines)
