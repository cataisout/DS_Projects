Uma possível estratégia é substituir com a média ou mediana, no entanto, parece que a distribuição não é gaussiana, ou seja, há uma diferença considerável entre a média e a mediana.
```python
demographic['MedianAgeInYears'].describe()
```

## Começando a análise

### Verificando os registros transacionais

A premissa inicial é encontrar três zipcodes mais viáveis para instalação de novas unidades da rede de laboratórios.

Primeiramente, vamos analisar a tabela transactional e observar as tendências de consumo, em relação a idade, gênero e localidade

```python
#criando coluna com idade 

transactional.head()
data_atual = datetime.today().date()
transactional['Age'] = data_atual.year - transactional['Date of birth'].dt.year
```

```python
transactional.head()
```

Visualizando consumo com base na idade, seguindo o mesmo range da tabela demographic

```python
# Definir os limites das faixas de idade
age_bins = [0, 5, 10, 15, 20, 25, 35, 45, 55, 60, 65, 75, 85, 120]
age_labels = ['Under 5', '5-9', '10-14', '15-19', '20-24', '25-34', '35-44', '45-54', '55-59', '60-64', '65-74', '75-84', '85 and Over']
# Criar as faixas de idade
transactional['Age Range'] = pd.cut(transactional['Age'], bins=age_bins, labels=age_labels)
```

Os três primeiros intervalos de idade são os que historicamente mais utilizam os serviços laboratoriais

```python
grouped_cost_age = transactional.groupby('Age Range')['Testing Cost'].sum()
grouped_cost_age = pd.DataFrame(grouped_cost_age)

grouped_cost_age['Testing Cost'] = grouped_cost_age['Testing Cost'].astype(np.float64)

grouped_cost_age = grouped_cost_age.sort_values(by='Testing Cost', ascending=False)
grouped_cost_age['Testing Cost'] = grouped_cost_age['Testing Cost'].map('{:,.2f}'.format)
grouped_cost_age
```

É possível observar que dentre os pacientes que mais consomem, a proporção de mulheres sempre é maior, mais ou menos o dobro

```python
grouped_counts = transactional.groupby('Age Range')['Gender'].value_counts()
grouped_counts = pd.DataFrame(grouped_counts)

grouped_gender_age = grouped_counts.loc[grouped_counts.index.get_level_values('Age Range').isin(['25-34', '35-44', '45-54'])]
grouped_gender_age
```

Abaixo temos os 3 laboratórios com os maiores faturamentos

```python
grouped_lab = pd.DataFrame(transactional.groupby('Lab Id')['Testing Cost'].sum())
grouped_lab = grouped_lab.sort_values(by='Testing Cost', ascending=False)
grouped_lab['Testing Cost'] = grouped_lab['Testing Cost'].map('{:,.2f}'.format)
grouped_lab.iloc[0:3]
```

Localização e Zipcode dos laboratórios de maior faturamento

```python
geocode[geocode['Lab Id'].isin(['L133', 'L139', 'L121'])]
```

### Procurando Localizações com base em dados econômicos e demográficos

Aqui, vamos buscar as áreas geográficas que mais se aproximam demograficamente e economicamente do nosso público alvo. Que como já definimos, são mulheres entre 25 e 54 anos. Tendo em vista que os grupos entre 35-44 são os com o pico de consumo

```python
filtered_demo = grouped_demographic[grouped_demographic['Demographic_Category'].isin(['Population_25to34Years', 'Population_35to44Years', 
                                                                     'Population_45to54Years', 'SexRatio(males per 100 females)', 'TotalPopulation'])]
filtered_demo
```

Ao analisar as faixas etárias mais significativas em termos de população, é estratégico concentrar o foco, a longo prazo, nas faixas etárias de 25 a 34 anos e de 35 a 44 anos. Isso se justifica pela queda acentuada na população após os 44 anos, o que impacta negativamente o volume de negócios, como visto anteriormente. 

Essa abordagem nos permite capitalizar em dois grupos distintos: o primeiro, entre 35 e 44 anos, responsável pelo maior faturamento; e o segundo, entre 25 e 34 anos, que não apenas contribui significativamente para o faturamento atual, mas também estabelece uma estratégia de longo prazo, aproveitando os três intervalos etários com maior potencial de consumo (25-34, 35-44, 45-54 anos). 

Dito isso, ainda assim o ideal é encontrar uma região que concentre o máximo possível de pessoas que estão dentro do público alvo

```python
top_areas = []
```

```python
filt_2534 = filtered_demo[filtered_demo.Demographic_Category == 'Population_25to34Years'].sort_values(by='Value', ascending=False)
top5Area_2534 = filt_2534.GeographicAreaName.iloc[0:10].tolist()
top_areas.append(top5Area_2534)
filt_2534.iloc[0:10]
```

```python
filt_3544= filtered_demo[filtered_demo.Demographic_Category == 'Population_35to44Years'].sort_values(by='Value', ascending=False)
top5Area_3544 = filt_3544.GeographicAreaName.iloc[0:10].tolist()
top_areas.append(top5Area_3544)
filt_3544.iloc[0:10]
```

```python
filt_4554= filtered_demo[filtered_demo.Demographic_Category == 'Population_45to54Years'].sort_values(by='Value', ascending=False)
top5Area_4554 = filt_4554.GeographicAreaName.iloc[0:10].tolist()
top_areas.append(top5Area_4554)
filt_4554.iloc[0:10]
```
