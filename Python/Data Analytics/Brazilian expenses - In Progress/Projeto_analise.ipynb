# Explorando os Dados

## Gasto Parlamentar por Região

```python
dados.head()
```

```python
grupo_por_regiao = dados.groupby(by ='Região')
gasto_por_regiao = grupo_por_regiao['Valor Líquido'].sum().sort_values(ascending = False)
pd.DataFrame(gasto_por_regiao)
```

```python

parlamentar_por_regiao = pd.crosstab(dados['Nome do Parlamentar'], dados['Região']).nunique().sort_values(ascending = False)
parlamentar_por_regiao
```

```python
gasto_parlamentar_por_regiao = (gasto_por_regiao / parlamentar_por_regiao).round(2).sort_values(ascending = False)
gasto_parlamentar_por_regiao = pd.DataFrame(gasto_parlamentar_por_regiao)
gasto_parlamentar_por_regiao.reset_index(inplace = True)
gasto_parlamentar_por_regiao.rename(columns = {0 : "Valor", 'index' : "Região"}, inplace = True)
gasto_parlamentar_por_regiao
# Nordeste possui maior gasto por parlamentar do Brasil 
```

```python
paleta = sns.color_palette('Blues_r')

ax = sns.barplot(data = gasto_parlamentar_por_regiao, x= 'Região', y = 'Valor', palette = paleta)
ax.figure.set_size_inches(10, 5)
```

## Explorando dados da Região Nordeste

```python
selecao = dados[dados['Região'] == 'Nordeste']
```

```python
#verificando gastos por partido 

gasto_por_partido = selecao.groupby(by = 'Partido')['Valor Líquido'].sum().sort_values(ascending = False)
qtd_parlamentar_por_partido = pd.crosstab(selecao['Nome do Parlamentar'], 
                                          selecao['Partido']).nunique()
gasto_por_partido = pd.DataFrame((gasto_por_partido / 
              qtd_parlamentar_por_partido).round(2).sort_values(ascending = False))
gasto_por_partido.reset_index(inplace = True)
gasto_por_partido.rename(columns = {0: "Valor"}, inplace = True)

ax = sns.barplot(data = gasto_por_partido, x= 'Partido', y = 'Valor')
ax.figure.set_size_inches(20, 10)
ax.set_title("Gasto por Partido - Nordeste", fontsize = 20)
```

```python
pd.DataFrame(selecao.groupby('Descrição')['Valor Líquido'].sum().sort_values(ascending = False))
```

```python
gastos_nordeste_por_uf = selecao.groupby(by = 'UF')['Valor Líquido'].sum().sort_values(ascending = False)
pd.DataFrame(gastos_nordeste_por_uf)
```

```python
dados_plot = pd.DataFrame(gastos_nordeste_por_uf)
dados_plot.reset_index(inplace = True)

ax = sns.barplot(data = dados_plot, x= 'UF', y = 'Valor Líquido', palette = paleta)
ax.ticklabel_format(style='plain', axis='y')
```

```python
parlamentares_UF_ne = pd.crosstab(selecao['Nome do Parlamentar'], 
                                  selecao['UF']).nunique().sort_values(ascending = False)
parlamentares_UF_ne
```

```python
gasto_medio_parlamentar_UF_ne = pd.DataFrame((gastos_nordeste_por_uf / 
                                 parlamentares_UF_ne).round(2).sort_values(ascending = False))
gasto_medio_parlamentar_UF_ne.rename(columns = {0: 'Valor'}, inplace = True)
gasto_medio_parlamentar_UF_ne.reset_index(inplace = True)
gasto_medio_parlamentar_UF_ne
```

```python
ax = sns.barplot(data = gasto_medio_parlamentar_UF_ne, x= 'UF', y = 'Valor', palette = paleta)

#Apesar de ter maior gasto, BA nao possui maior gasto por parlamentar
```

```python
ax = sns.boxplot(data = selecao, x = 'Valor Líquido', y='UF' )
```

```python
#Verificando distribuição no Ceará
selecao = dados[dados['UF'] == 'CE']

grupo = selecao.groupby(by = 'Partido')

valor_por_partido = grupo['Valor Líquido'].sum().sort_values(ascending = False)
qtd_parlamentar_por_partido = selecao.value_counts('Partido')
valor_parlamentar_por_partido = pd.DataFrame(((valor_por_partido / 
  qtd_parlamentar_por_partido).round(2)).sort_values(ascending = False))
valor_parlamentar_por_partido
valor_parlamentar_por_partido.reset_index(inplace = True)
valor_parlamentar_por_partido.rename(columns = {0 : 'Valor'}, inplace = True)
```

```python
ax = sns.barplot(data = valor_parlamentar_por_partido.head(), x= 'Partido', y = 'Valor', palette = paleta)
ax.figure.set_size_inches(10, 6)
ax.set_title("TOP 5 - Gasto Médio por Parlamentar por Partido - Ceará", fontsize = 12)
```

### Verificando partidos com mais outliers

```python

pdt = selecao[selecao['Partido'] == 'PDT']
pros = selecao[selecao['Partido'] == 'PROS']
psl = selecao[selecao['Partido'] == 'PSL']
```

```python
pdt['Valor Líquido'].describe()
```

```python
pdt['Valor Líquido'].mode()
```

```python
#desobrindo outliers
q1 = pdt['Valor Líquido'].quantile(q=0.25)
q3 = pdt['Valor Líquido'].quantile(q=0.75)
aiq = q3 - q1


Limite_Superior = q3 + (1.5 * aiq) 
Limite_Inferior = q1 - (1.5 * aiq)
```

```python
pdt[pdt['Valor Líquido'] >= Limite_Superior].head()
```
